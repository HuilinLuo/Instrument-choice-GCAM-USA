```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
import folium as fo
import plotly as ply
import plotly.graph_objects as go
import matplotlib

from matplotlib.gridspec import GridSpec
from matplotlib.ticker import FuncFormatter
import matplotlib as mpl

import seaborn as sns
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
import matplotlib
```

```python
df = pd.read_excel("tech costs.xlsx", sheet_name ="summary_plot")
default = df[df['type']=='Default']
inno =df[df['type']=='Innovation']
```

```python

mpl.rcParams['font.family'] = 'Arial'


blue_colors = ['#1f77b4', "#1E96A6", "#B5D9D9"]   
brown_colors = ['#4C3329','#964B00'] 
colors = blue_colors + brown_colors  

fig, axs = plt.subplots(1, 2, figsize=(12, 6), sharey=True)


for i, column in enumerate(default.columns[2:]): 
    axs[0].plot(default['Year'], default[column], label=column, 
                color=colors[i]
               )  

axs[0].set_title('(a) Default innovation', fontsize=16)
axs[0].set_xlabel('Year', fontsize=14)
axs[0].set_ylabel('Percent changes in capital costs relative to 2015', fontsize=14)
axs[0].legend(title='Technologies', fontsize=10)

percent_formatter = lambda x, _: f"{x:.0%}"
axs[0].yaxis.set_major_formatter(FuncFormatter(percent_formatter))


for i, column in enumerate(inno.columns[2:]):  
    axs[1].plot(inno['Year'], inno[column], label=column, color=colors[i ]) 

axs[1].set_title('(b) Accelerated innovation', fontsize=16)
axs[1].set_xlabel('Year', fontsize=14)

axs[1].legend(title='Technologies', fontsize=10)

plt.tight_layout()

plt.show()
```
